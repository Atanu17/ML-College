```python
null_var = df.isnull().sum()/df.shape[0]*100 # percentage of null values corresponding to columns
drop_columns = null_var[null_var > 17].keys()
drop_columns
```

```python
df2_drop_col = df.drop(drop_columns, axis = 1) # dropping the columns
df2_drop_col.shape
```

```python
df3_drop_rows = df2_drop_col.dropna()
df3_drop_rows.shape
```

```python
sns.heatmap(df3_drop_rows.isnull())
plt.show()
```
