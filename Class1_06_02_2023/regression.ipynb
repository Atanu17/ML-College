```python
# func: linregress(x,y) -> slope, intercept, r, p, std_err

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats

x=np.random.normal(0,10.0,1000)
y=np.random.normal(0,10.0,1000)

slope, intercept, r, p, std_err = scipy.stats.linregress(x,y)


# calculate the speed of the car
def f(x):
    return slope*x + intercept
mymodel = list(map(f, x))

plt.figure(figsize=(10,10))
plt.scatter(x,y)
plt.plot(x, mymodel)
plt.show()

print(f(10))
```

```python
# polynomial regression
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats

x = [ 1, 2, 3, 5, 6, 7, 8, 9, 10, 12, 13, 14, 15, 16, 18, 19, 21, 22]
y = [100, 90, 80, 60, 55, 55, 45, 79, 70, 80, 78, 89, 87, 90, 90, 30, 15, 34]

print(len(x),len(y))

mymodel = np.poly1d(np.polyfit(x,y,3))
myline = np.linsplace(1, 22, 100)

plt.scatter(x,y)
plt.plot(myline, mymodel(myline))

plt.show()
```
