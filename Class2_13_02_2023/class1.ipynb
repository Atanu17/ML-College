```python
null_var = df.isnull().sum()/df.shape[0]*100 # percentage of null values corresponding to columns
drop_columns = null_var[null_var > 20].keys()
drop_columns
```

```python
df_num = df.drop(drop_columns, axis = 1) # dropping the columns
df_num.shape
```

```python
df_new = df_num.select_dtypes(include = ['int64', 'float64'])
missing_num_var =  [var for var in df_new.columns if df_new[var].isnull().sum()>0]
missing_num_var
```

```python
# data dist for all integer and float columns
plt.figure(figsize = (10, 10))
for i, col in enumerate(missing_num_var):
        plt.subplot(2, 2, i+1)
        sns.distplot(df_new[col].fillna(0), bins = 20) # fillna for this version only
plt.show()
```

#### Replacing numerical null values with Mean or Median of the data values
- For more outliers in dataset we use mean
- For skewed graph we use median

```python
df_mean = df_new.fillna(df_new.mean())
df_median = df_new.fillna(df_new.median())
df_mode = df_new.fillna(df_new.mode())

plt.figure(figsize = (10, 10))
for i, col in enumerate(missing_num_var):
        plt.subplot(2, 2, i+1)
        sns.distplot(df_new[col].fillna(0), bins = 20) # fillna for this version only
        sns.distplot(df_mean[col], bins = 20)
plt.show()
```
