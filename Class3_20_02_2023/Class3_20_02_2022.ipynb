```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/ML/Class3/Bengaluru_House_Data.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['size'] = df['size'].apply(lambda x: str(x).split(" ")[0])
```

```python
df = df.drop("society", axis = 1)
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.drop([568], axis = 0)
df.head()
df.isnull().sum()
```

```python
mode_val = df["bath"].mode()[0]
mode_val
```

```python
df['bath'] = df['bath'].fillna(mode_val)
df.isnull().sum()
```

```python
df['balcony'] = df['balcony'].fillna(df["balcony"].mode()[0])
df.isnull().sum()
```

```python
df['total_sqft'] = df['total_sqft'].apply(lambda x: str(float(x.split("Sq. Meter")[0]) * 10.7639) if "Sq. Meter" in x else x)
```

```python
df.iloc[410]
```

```python
df['total_sqft'] = df['total_sqft'].apply(lambda x: str(float(x.split("Perch")[0]) * 272.25) if "Perch" in x else x)
df['total_sqft'] = df['total_sqft'].apply(lambda x: str(float(x.split("Sq. Yards")[0])*9.0) if "Sq. Yards" in x else x)
df['total_sqft'] = df['total_sqft'].apply(lambda x: str(float(x.split("Acres")[0])*43560.0) if "Acres" in x else x)
df['total_sqft'] = df['total_sqft'].apply(lambda x: str(float(x.split("Cents")[0])*435.56) if "Cents" in x else x)
df['total_sqft'] = df['total_sqft'].apply(lambda x: str(float(x.split("Guntha")[0])*1089.0) if "Guntha" in x else x)
df['total_sqft'] = df['total_sqft'].apply(lambda x: str(float(x.split("Grounds")[0])*2400.0) if "Grounds" in x else x)
```

```python
df["total_sqft"] = df["total_sqft"].apply(lambda x: (float(x.split("-")[0]) + float(x.split("-")[1]))/2.0 if "-" in x else x)
```

```python
df['size'] = df['size'].astype("float") + df["bath"] + df["balcony"]
df = df.drop(columns = ["bath", "balcony"], axis = 1)
```

```python
df = df[df["size"].notnull()]
df["size"] = df["size"].astype("int64")
df.head(15)
```
